Jetzt vergleichen wir den Einfluss der unterschiedlichen Werte von $b$ und $k$.

dazu mssen die anderen Datenstze eingelesen und mit in eine Grafik geplottet werden.

```python
data1=pd.read_csv('Federschwinger_k=100000_b=30.csv')
data1=data1.set_axis(['t[s]','x[m]','F[N]'], axis='columns')
data1['x[mm]']=1000*data1['x[m]']
```

```python
data2=pd.read_csv('Federschwinger_k=100000_b=100.csv')
data2=data2.set_axis(['t[s]','x[m]','F[N]'], axis='columns')
data2['x[mm]']=1000*data2['x[m]']
```

Immer wenn exakt die gleiche Aufgabe mehrfach ausgefhrt wird lohnt sich die Definition einer Funktion:

```python
def Datensatzlesenundvorbereiten(filename):
    data=pd.read_csv(filename)
    data=data.set_axis(['t[s]','x[m]','F[N]'], axis='columns')
    data['x[mm]']=1000*data['x[m]']
    return data
```

```python
data3=Datensatzlesenundvorbereiten('Federschwinger_k=100000_b=0.csv')
```

```python
plt.plot(data1['t[s]'],data1['x[mm]'])
plt.plot(data2['t[s]'],data2['x[mm]'])
plt.plot(data3['t[s]'],data3['x[mm]'])
plt.xlabel('$t$ in s')
plt.ylabel('$x$ in mm')
plt.xlim([0,10])
```

Schrnke den Bereich auf 5s ein und fge eine Legende hinzu

```python
plt.plot(data1['t[s]'],data1['x[mm]'],label='k=100000_b=30')
plt.plot(data2['t[s]'],data2['x[mm]'],label='k=100000_b=100')
plt.plot(data3['t[s]'],data3['x[mm]'],label='k=100000_b=0')
plt.xlabel('$t$ in s')
plt.ylabel('$x$ in mm')
plt.legend()
plt.xlim([0,5])
```
