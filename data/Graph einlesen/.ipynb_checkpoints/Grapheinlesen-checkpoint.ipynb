Einlesen der Importe

```python
df_import=pd.read_csv('imports.csv',sep=';', decimal=',',names=['year','price']).sort_values('year')
df_import.head()
```

nochmal sauberes Einlesen der Exporte

```python
df_export=pd.read_csv('exports.csv',sep=';', decimal=',',names=['year','price']).sort_values('year')
df_export.head()
```

# graphische Darstellung

```python
plt.plot(df_import.year, df_import.price)
plt.plot(df_export.year, df_export.price)
```

```python
plt.plot(df_import.year, df_import.price,label='import')
plt.plot(df_export.year, df_export.price,label='export')
plt.legend()
```

```python
plt.plot(df_import.year, df_import.price,label='import')
plt.plot(df_export.year, df_export.price,label='export')
plt.xlabel('year')
plt.ylabel('price in £')
plt.legend()
```

```python
plt.plot(df_import.year, df_import.price,label='import')
plt.plot(df_export.year, df_export.price,label='export')
plt.fill_between(df_export.year,df_export.price,color='darkred')
plt.fill_between(df_import.year,df_import.price,color='white')
plt.xlabel('year')
plt.ylabel('price in £')
plt.legend()
```
